###1 Use the following code for making database connection to Microsoft SQL Serve

````python
import pymssql
import pandas as pd
conn = pymssql.connect(server='SQL6020.site4now.net', database='db_a34f35_coded' ,user='db_a34f35_coded_admin', password='coded123*')
````

###2 Select the appropriate JOIN clause to retrieve all records from the Orders and Customers tables where there is a matching entry in both tables.

````python
SQL_statement="SELECT Orders.*, Customers.* FROM Orders INNER JOIN Customers ON Orders.CustomerID = Customers.CustomerID"
df_customers = pd.read_sql_query(SQL_statement, conn)
df_customers
````

###3 Select the appropriate JOIN clause to retrieve all records from the Orders and Employees tables where there is a matching entry in both tables.

````python
SQL_statement="SELECT Orders.*, Employees.* FROM Orders INNER JOIN Employees ON Orders.EmployeeID = Employees.EmployeeID"
df_Employees = pd.read_sql_query(SQL_statement, conn)
df_Employees
````

###4Select the appropriate JOIN clause to retrieve all records from the Customers table along with any corresponding records from the Orders table.

````python
SQL_statement="SELECT Orders.*, Customers.* FROM Customers LEFT JOIN Orders ON Orders.CustomerID = Customers.CustomerID"
df_customers = pd.read_sql_query(SQL_statement, conn)
df_customers
````

###5 Select the appropriate JOIN clause to retrieve all records from the OrderDetails table along with any corresponding records from the Products table.

````python
SQL_statement="SELECT OrderDetails.*,  Products.* FROM OrderDetails LEFT JOIN Products ON Products.ProductID = OrderDetails.ProductID"
df_OrderDetails = pd.read_sql_query(SQL_statement, conn)
df_OrderDetails
````

###6 Select the appropriate JOIN clause to retrieve all records from the Products, Categories, and Suppliers tables where there is a matching entry in three tables.

````python
SQL_statement = "SELECT Products.*, Categories.*, Suppliers.* FROM Products INNER JOIN Categories ON Products.CategoryID = Categories.CategoryID INNER JOIN Suppliers ON Products.SupplierID = Suppliers.SupplierID"
df_customers = pd.read_sql_query(SQL_statement, conn)
df_customers
````
